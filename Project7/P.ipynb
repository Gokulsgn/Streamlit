Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection and Processing

```python
# loading the data from csv file to pandas dataframe
car_dataset = pd.read_csv(r'C:\Users\gokul\Documents\GitHub\Machine_Learning\Project7\car data.csv')
```

```python
# inspecting the first 5 rows of the dataframe
car_dataset.head()
```

```python
# checking the number of rows and columns
car_dataset.shape
```

```python
# getting some information about the dataset
car_dataset.info()
```

```python
# checking the number of missing values
car_dataset.isnull().sum()
```

```python
# checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the Categorical Data

```python
#Encoding "Fuel_Type" column
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2},'Seller_Type':{'Dealer':0,'Individual':1},'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_dataset.head()
```

Splitting the data and Target

```python
X = car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```


Model Training

Linear Regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Prediction on Training Data
training_prediction = lin_reg_model.predict(X_train)
```

```python
# R Squared Error 

error_score = metrics.r2_score(Y_train,training_prediction)
print("R Squared Error :",error_score)
```

Visualize teh actual prices and predicted prices

```python
plt.scatter(Y_train,training_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted prices")
plt.show()
```

```python
#Prediction on Test Data
test_prediction = lin_reg_model.predict(X_test)
```

```python
# R Squared Error 

error_score = metrics.r2_score(Y_test,test_prediction)
print("R Squared Error :",error_score)
```

```python
plt.scatter(Y_test,test_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted Price")
plt.title("Actual price vs predicted prices")
plt.show()
```
